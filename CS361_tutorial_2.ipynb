# What is the multiplication rule + an example

From the textbook, the definition/formula 3.3 states that two events _A_ and _B_ are independent if and only if:

$$P(A \cap B) = P(A) * P(B) $$

When you want to calculate the probability of two, independent events happening at the same time, you can use this rule, called the *multiplication rule*.

## Example
In a game of dice what is the probability of snake eyes? (You roll two dice and both land on a 1)



```python
import itertools

dice1 = [1, 2, 3, 4, 5, 6]
dice2 = [1, 2, 3, 4, 5, 6]

c = list(itertools.product(dice1, dice2))

print(c)
print("There are:", len(c), "possible pairs")
```

By the definition of probability:

$$P(\text{snake eyes}) = \frac{\text{successes}}{\text{possible outcomes}} = \frac{1}{36} $$

Similarly notice that the probability of dice 1 rolling a 1 is $\frac{1}{6}$ and dice 2 rolling a 1 is $\frac{1}{6}$

which, using the formula is: 

$$ P(\text{Dice 1 Rolling a 1}) * P(\text{Dice 2 Rolling a 2}) = \frac{1}{6} * \frac{1}{6} = \frac{1}{36} = P(\text{Dice 1 Rolling a 1} \cap \text{Doce 2 Rolling a 2}) = P(\text{snake eyes})   $$


#### Note: If the events are NOT independent, the multiplication rules does NOT apply

In that case, you would use another formula learned in class: 

$$ P(A \cap B) = P(A) * P(B|A)$$

where $P(B|A)$ means the probability of B given A.
