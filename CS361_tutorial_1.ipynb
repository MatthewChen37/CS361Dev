# Why do we need to standardize data?

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
```

### Let's say we ask two different groups of people what their favorite numbers are...

```python
dataset1_x = np.asarray([-7, -6, -5, -4, -3, -2, -1])
dataset1_y = np.asarray([1, 2, 3, 4, 3, 2, 1])

dataset2_x = np.asarray([-2.5, -1, 0.5, 2, 3.5, 5, 6.5])
dataset2_y = np.asarray([1, 2, 3, 4, 3, 2, 1])
```

```python
plt.bar(dataset2_x, dataset2_y, color="red", label="Group 2")
plt.bar(dataset1_x, dataset1_y, color="green", label="Group 1")
plt.legend()
plt.title("Favorite Number")
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.show()
```

### Now let's look at the them standardized...

```python
dataset1_x_standardized = (dataset1_x - np.mean(dataset1_x)) / np.std(dataset1_x)
dataset2_x_standardized = (dataset2_x - np.mean(dataset2_x)) / np.std(dataset2_x)
```

```python
plt.bar(dataset2_x_standardized, dataset2_y, color="red")
plt.title("Favorite Number")
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.show()

plt.bar(dataset1_x_standardized, dataset2_y, color="green")
plt.title("Favorite Number")
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.show()
```
